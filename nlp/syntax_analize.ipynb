```python
from natasha import Segmenter, NewsEmbedding, NewsSyntaxParser, Doc
```

```python
emb = NewsEmbedding() #embedder
segmenter = Segmenter() #tokenizer
syntax_parser = NewsSyntaxParser(emb) #parser
```

```python
with open('asya_1858.txt', 'r') as file:
    text = file.readlines()

text
```

```python
# Не забыть, что мы работаем по предложениям, а не с текстом целиком
doc = Doc(text[5])
doc.segment(segmenter)
for token in doc.tokens:
   print(token)

doc
```

```python
doc.parse_syntax(syntax_parser)
for token in doc.tokens:
    print(token)
```
