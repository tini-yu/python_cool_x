```python
class predobrabotka:

    def __init__(self, text: str):
        self.text = text

    def PreProcessing(self):
        doc = nlp(text)
        words_filtered = [token.text for token in doc if not token.is_punct and not token.is_stop]
        for el in words_filtered:
            if ' ' in el or '\n' in el or '\xa0' in el:
                words_filtered.remove(el)
            if 'V' in el or 'X' in el or 'I' in el:
                words_filtered.remove(el)
        
        
```
