```python
import gensim.downloader
import spacy
nlp = spacy.load("ru_core_news_sm")
from collections import Counter
import nltk
import pandas as pd
```

```python
word2vec_rus = gensim.downloader.load('word2vec-ruscorpora-300')
```

```python
with open('burmistr_1847.txt', 'r', encoding='utf8') as file:
    text = file.read()

text
```

```python
doc = nlp(text)
wf = [token.text for token in doc if not token.is_punct and not token.is_stop]
removal = [' ', '\n', '\xa0', 'note']
for x in range(len(wf)):
    for y in removal:
        if y in wf:
            wf.remove(y)



    # if ' ' in el or '\n' in el or '\xa0' in el:
    #     words_filtered.remove(el)
    # if 'V' in el or 'X' in el or 'I' in el or 'note' in el:
    #     words_filtered.remove(el)

wf
```

```python
counter = Counter(wf)

# most_common = counter.most_common(1)[0][0]
# print(most_common)  # 👉️ one

most_common_100 = counter.most_common(100)
most_common_100 = most_common_100[0:100]
print(most_common_100)
```

```python
len(wf), len(most_common_100)
```

```python
df = pd.DataFrame(most_common_100)
words = list(df[0])
# df.nlargest(100, 1)
words
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
for word in words:
    #print(morph.parse(word))
    print(morph.parse(word)[0].tag.POS) # [0], т.к. берем первое значение слова
```
